```python
#Install Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
import hvplot.pandas
```

```python
# Ticker List
ticker_list = ['^GSPC',
'^IXIC',
'^DJI']
```

```python
#Creates a dataframe of the closing price using the ticker_list starting from 1997
x = 0
while x in range(len(ticker_list)):
    if x == 0:
        df = yf.Ticker(ticker_list[x]).history(start='1997-01-01')['Close'].rename(ticker_list[x])
    else:
        df1 = yf.Ticker(ticker_list[x]).history(start='1997-01-01')['Close'].rename(ticker_list[x])
        df = pd.concat([df, df1], axis=1)
    x += 1
df.head()
```

```python
#Select Ticker to use for Fibonacci Retracement
DJI = "^DJI"
ticker = yf.Ticker(DJI)
df_1 = ticker.history(interval="1d",start="1997-01-01",end="2022-11-04")
df_1.head()

#Sets highest and lowest swings equal to respective closing dates for Fibonacci Retracement
highest_swing = -1
lowest_swing = -1
for i in range(1,df_1.shape[0]-1):
    if df_1['High'][i] > df_1['High'][i-1] and df_1['High'][i] > df_1['High'][i+1] and (highest_swing == -1 or df_1['High'][i] > df_1['High'][highest_swing]) :
        highest_swing = i
    if df_1['Low'][i] < df_1['Low'][i-1] and df_1['Low'][i] < df_1['Low'][i+1] and (lowest_swing == -1 or df_1['Low'][i] < df_1['Low'][lowest_swing]) :
        lowest_swing = i

#Creates Fibonacci Retracement Lines at respective percentages
ratios = [0,0.236, 0.382, 0.5 , 0.618, 0.786,1]
colors = ["black","r","g","b","cyan","magenta","yellow"]
levels = []
max_level = df_1['High'][highest_swing]
min_level = df_1['Low'][lowest_swing]
for ratio in ratios:
    if highest_swing > lowest_swing: # Uptrend
        levels.append(max_level - (max_level-min_level)*ratio)
    else: # Downtrend
        levels.append(min_level + (max_level-min_level)*ratio)

plt.rcParams['figure.figsize'] = [20,10]
plt.rc('font', size=14)
plt.plot(df_1['Close'])
start_date = df_1.index[min(highest_swing,lowest_swing)]
end_date = df_1.index[max(highest_swing,lowest_swing)]
print(start_date)
for i in range(len(levels)):
# Places the Fibonacci lines on the graph
    plt.hlines(levels[i],df_1.index[0], df_1.index[-1],label="{:.1f}%".format(ratios[i]*100),colors=colors[i], linestyles="dashed")

plt.legend()
plt.show()
```
